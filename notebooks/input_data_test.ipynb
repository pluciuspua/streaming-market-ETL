```python
import re
from typing import Iterable, List, Tuple, Optional
def sanitize_texts(texts: Iterable, escape_double: bool = False, max_len: Optional[int] = None) -> List[str]:
    """
    Ensure every item is a safe string:
    - Convert non-strings to str
    - Strip surrounding whitespace
    - Normalize “smart” quotes to straight quotes
    - Remove control characters that can break JSON or editors
    - Optionally escape double quotes (useful if you need to paste the value into a double-quoted literal)
    - Optionally enforce max_len
    """
    out = []
    for t in texts if not isinstance(texts, str) else [texts]:
        if t is None:
            t = ""
        if not isinstance(t, str):
            t = str(t)
        t = t.strip()
        # Normalize smart quotes
        t = t.replace("\u201c", '"').replace("\u201d", '"').replace("\u2018", "'").replace("\u2019", "'")
        # Remove control characters except common whitespace
        t = re.sub(r'[\x00-\x08\x0b\x0c\x0e-\x1f\x7f]', '', t)
        if escape_double:
            t = t.replace('"', r'\"')
        if max_len is not None and len(t) > max_len:
            t = t[:max_len]
        # Ensure JSON serializable (fallback to unicode-escaped if necessary)
        try:
            json.dumps(t)
        except (TypeError, ValueError):
            t = t.encode('unicode_escape').decode('ascii')
        out.append(t)
    return out

url = " https://finbert-api-196202225837.asia-southeast1.run.app/batch_predict"
texts = [
    ''''"NEW YORK CITY and NEW ORLEANS, Sept. 17, 2025 ( GLOBE NEWSWIRE ) -- Former Attorney General of Louisiana, Charles C. Foti, Jr., Esq., a partner at the law firm of Kahn Swick & Foti, LLC ( "KSF" ) , announces that KSF has commenced an investigation into Apple Inc. ( "Apple" or the "Company" ) AAPL."''',
    ''''"NEW YORK CITY and NEW ORLEANS, Sept. 17, 2025 ( GLOBE NEWSWIRE ) -- Former Attorney General of Louisiana, Charles C. Foti, Jr., Esq., a partner at the law firm of Kahn Swick & Foti, LLC ( "KSF" ) , announces that KSF has commenced an investigation into Apple Inc. ( "Apple" or the "Company" ) AAPL."''',
    ''''"NEW YORK CITY and NEW ORLEANS, Sept. 17, 2025 ( GLOBE NEWSWIRE ) -- Former Attorney General of Louisiana, Charles C. Foti, Jr., Esq., a partner at the law firm of Kahn Swick & Foti, LLC ( "KSF" ) , announces that KSF has commenced an investigation into Apple Inc. ( "Apple" or the "Company" ) AAPL."''',
    ''''"NEW YORK CITY and NEW ORLEANS, Sept. 17, 2025 ( GLOBE NEWSWIRE ) -- Former Attorney General of Louisiana, Charles C. Foti, Jr., Esq., a partner at the law firm of Kahn Swick & Foti, LLC ( "KSF" ) , announces that KSF has commenced an investigation into Apple Inc. ( "Apple" or the "Company" ) AAPL."''',
    ''''"NEW YORK CITY and NEW ORLEANS, Sept. 17, 2025 ( GLOBE NEWSWIRE ) -- Former Attorney General of Louisiana, Charles C. Foti, Jr., Esq., a partner at the law firm of Kahn Swick & Foti, LLC ( "KSF" ) , announces that KSF has commenced an investigation into Apple Inc. ( "Apple" or the "Company" ) AAPL."''',
    ''''"NEW YORK CITY and NEW ORLEANS, Sept. 17, 2025 ( GLOBE NEWSWIRE ) -- Former Attorney General of Louisiana, Charles C. Foti, Jr., Esq., a partner at the law firm of Kahn Swick & Foti, LLC ( "KSF" ) , announces that KSF has commenced an investigation into Apple Inc. ( "Apple" or the "Company" ) AAPL."''',
    ]
texts = sanitize_texts(texts)
payload = {"texts": texts
}

response = requests.post(url, json=payload)
print(response.json())
```

```python
SYMBOL = "MSFT"

def fetch_stock_price(symbol):
    url = f"https://www.alphavantage.co/query"
    params = {
        "function": "NEWS_SENTIMENT",
        "tickers": symbol,
        "apikey": API_KEY,
        "sort": "RELEVANCE",
        "time_from":"20250914T0000"
    }
    r = requests.get(url, params=params).json()
    return r


with open("test_news_data.json", "w") as file:
    r = fetch_stock_price(SYMBOL)
    json.dump(r, file, indent=4)
    pprint.pprint(r)
```
